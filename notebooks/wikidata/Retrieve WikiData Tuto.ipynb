
# 2. Use requests

Wikipedia documentation API:
https://www.mediawiki.org/wiki/API:Main_page


```python
import requests
from bs4 import BeautifulSoup
```

## Use query

```python
S = requests.Session()

URL = "https://en.wikipedia.org/w/api.php"

PARAMS = {'action': 'query',
          'prop': 'extracts', 
          'explaintext': '',
          'rvprop': 'ids', 
          'titles': query,
          'format': 'json'
}

R = S.get(url=URL, params=PARAMS)
DATA = R.json()['query']['pages']
soup = BeautifulSoup(DATA[[k for k in DATA.keys()][0]]['extract'], "html.parser")

html_text = soup.get_text()
print(html_text)
```

## Use parse

```python
S = requests.Session()

URL = "https://en.wikipedia.org/w/api.php"

PARAMS = {
    "action": "parse",
    "page": query,
    "format": "json"
}

R = S.get(url=URL, params=PARAMS)
DATA = R.json()["parse"]["text"]["*"]
soup = BeautifulSoup(DATA, "html.parser")

html_text = soup.get_text()
print(html_text)
```
